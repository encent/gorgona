# Generate bacteria-only environment

```python
# Define and initialize parameters
params = {
	"input_file_path": "data/01_bacteria_amount/bac_amount_workshop.tsv",
	"generation_type": "bacteria",
	"mode": "hybrid",
	"bottom_padding": 0,
	"xlim": (0, 1), # (0, 20)
	"ylim": (0, 1), # (0, 20)
	"zlim_villi": (0, 1), # (0, 2.5)
	"zlim_bacteria": (0, 1), # (0, 2.5)
	"bsize": 0.04, # 0.04 or 0.12 for test
	"bmass": 0.000064, # 0.000064 or 0.001728 for test
    # Add more parameters if needed
}

# Create an instance of BacteriaVilliCoordinatesGenerator
bacteria_villi_generator = BacteriaVilliCoordinatesGenerator(**params)

# Call the generate_coordinates method
bacteria_villi_generator.generate_coordinates(output_dir_path="data/02_coordinates/", 
											  plot_figs=True, 
											  figures_path="figures")
```
