```python
from gorgona.analyzer import Heatmaps, Curves
```

### Bacteria gt

```python
heatmaps = Heatmaps(experiment_id="data/03_simulation", 
					experiment_type="bacteria_gt", 
					generation_type="bacteria", 
					cluster_cutoff=0.06, 
					start_frame=1500, 
					end_frame=2000, 
					delta=50, 
					tol=1e-5, 
					input_bac_amount_file_path="data/01_bacteria_amount/bac_amount_workshop.tsv", 
					input_func_distances_file_path="data/input_files/genome_functional.distances.txt",
					)
```

```python
heatmaps.plot_observed(vmax=0.15, vmin=-0.95)
```

```python
heatmaps.plot_expected(vmax=0.3, vmin=0.1)
```

```python
heatmaps.plot_diff(vmax=2, vmin=-2)
```

```python
heatmaps.plot_corr()
```

```python
curves = Curves(
	experiment_id="data/03_simulation/", 
	experiment_type="bacteria_gt", 
)
```

```python
curves.plot_main(ymin=None, ymax=None, xmax=0.03 * 10e7)
```

```python
curves.plot_bacteria(ymin=None, ymax=None, xmax=0.03 * 10e7)
```

```python
curves.plot_energy(ymin=None, ymax=None, xmax=0.03 * 10e7)
```

```python
curves.plot_villi(ymin=None, ymax=None, xmax=0.03 * 10e7)
```
