# Run simulations

```python
import subprocess
```

### Parameters

```python
n_steps=2000000
id_bac_coef=0.00000133649289
vil_bac_coef=0.00000133649289
bac_size=0.04
bac_nve_limit=0.4375
id_bac_cf=0.06#0.4375
diff_bac_cf=0.06#0.4375
vil_bac_cf=0.35
global_cf=0.35
neighbor_skin_distance=0.35 #0.8
neigh_modify_size=10000
page_size=100000
vil_bac_size=2.02
bac_veloc=4e-09
temp_bac=4e-09
```

### Bacteria, gt

```python
cmd = [
    "mpirun", "-np", "4", "lmp_mpi",
    "-in", "run.in",
    "-var", "n_steps", str(n_steps),
    "-var", "temp_bac", str(temp_bac),
    "-var", "bac_nve_limit", str(bac_nve_limit),
    "-var", "bac_veloc", str(bac_veloc),
    "-var", "neigh_modify_size", str(neigh_modify_size),
    "-var", "bac_size", str(bac_size),
    "-var", "id_bac_cf", str(id_bac_cf),
    "-var", "vil_bac_size", str(vil_bac_size),
    "-var", "vil_bac_cf", str(vil_bac_cf),
    "-var", "id_bac_coef", str(id_bac_coef),
    "-var", "vil_bac_coef", str(vil_bac_coef),
    "-var", "global_cf", str(global_cf),
    "-var", "diff_bac_cf", str(diff_bac_cf),
    "-var", "page_size", str(page_size),
    "-var", "neighbor_skin_distance", str(neighbor_skin_distance)
]

print("Executing:", " ".join(cmd))
subprocess.run(cmd, cwd="data/03_simulation/bacteria_gt")
```

### Bacteria, reshuffled

```python
cmd = [
    "mpirun", "-np", "4", "lmp_mpi",
    "-in", "run.in",
    "-var", "n_steps", str(n_steps),
    "-var", "temp_bac", str(temp_bac),
    "-var", "bac_nve_limit", str(bac_nve_limit),
    "-var", "bac_veloc", str(bac_veloc),
    "-var", "neigh_modify_size", str(neigh_modify_size),
    "-var", "bac_size", str(bac_size),
    "-var", "id_bac_cf", str(id_bac_cf),
    "-var", "vil_bac_size", str(vil_bac_size),
    "-var", "vil_bac_cf", str(vil_bac_cf),
    "-var", "id_bac_coef", str(id_bac_coef),
    "-var", "vil_bac_coef", str(vil_bac_coef),
    "-var", "global_cf", str(global_cf),
    "-var", "diff_bac_cf", str(diff_bac_cf),
    "-var", "page_size", str(page_size),
    "-var", "neighbor_skin_distance", str(neighbor_skin_distance)
]

print("Executing:", " ".join(cmd))
subprocess.run(cmd, cwd="data/03_simulation/bacteria_reshuffled")
```
