```python
from gorgona.analyzer import Heatmaps, Curves
```

### Both gt

```python
heatmaps = Heatmaps(experiment_id="03_run_experiments_25-06-24_vilbac_0.04_75K", 
					experiment_type="both_gt", 
					generation_type="both", 
					cluster_cutoff=0.06, 
					start_frame=2500, 
					end_frame=3000, 
					delta=50, 
					tol=1e-5, 
					input_bac_amount_file_path="../data/01_bacteria_amount/bac_amount_76287_mult_63.97465212588298_thresh_10.tsv", 
					input_func_distances_file_path="../data/input_files/genome_functional.distances.txt",
					)
```

```python
heatmaps.plot_observed(vmax=0.15, vmin=-0.95)
```

```python
heatmaps.plot_expected(vmax=0.3, vmin=0.1)
```

```python
heatmaps.plot_diff(vmax=2, vmin=-2)
```

```python
heatmaps.plot_corr()
```
